```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab/StudentsPerformance.csv")
df.head()
```

```python
sns.catplot(data=df,x='gender', y='math score',kind='box',row='lunch')
```

```python
sns.catplot(data=df,x='gender', y='math score',kind='box',col='lunch')
```

```python
sns.catplot(data=df,x='gender', y='math score',kind='box',
            col='lunch',row='test preparation course')
```

```python
sns.catplot(data=df,x='gender', y='math score',kind='violin',
            col='lunch',row='test preparation course')
```

```python
sns.catplot(data=df,x='gender', y='math score',kind='bar',
            col='lunch',row='test preparation course')
```

```python
sns.catplot(data=df,x='gender', y='math score',kind='point',
            col='lunch',row='test preparation course')
```

```python
sns.catplot(data=df, y='math score',kind='count',
            col='lunch',row='test preparation course')
```

```python
g = sns.PairGrid(data=df)
g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.kdeplot)
g = g.map_lower(sns.kdeplot)
```

```python
g = sns.PairGrid(data=df,hue='gender')
g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.kdeplot)
g = g.map_lower(sns.kdeplot)
g = g.add_legend()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab/dm_office_sales.csv")
df.head()
```

```python
plt.figure(figsize=(8,5),dpi=80)
sns.rugplot(x='salary', data=df,height=0.5)
```

```python
# displot no se debe usar porque va a ser deprecado
sns.set(style='darkgrid')
sns.displot(data=df, x='salary',kde=True,rug=True,bins=40, color='orange', edgecolor='green',
            linewidth=3)
```

```python
sns.histplot(data=df, x='salary',kde=True)
```

```python
sns.kdeplot(data=df, x='salary')
```
